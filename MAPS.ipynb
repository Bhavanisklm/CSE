```python
coords_1=[[16.780001,80.849998],
          [16.7430141,80.8668691],
          [16.703285, 81.100388],
          [17.004393,81.783325],
          [ 17.0504374,82.1659243],
          [16.989065,82.247467]]
          
import folium
from branca.element import Figure
fig5=Figure(height=550,width=750)
map5=folium.Map(location=[16.780001,80.849998],zoom_start=7)
fig.add_child(map1)
map1
```

```python
f1=folium.FeatureGroup('route')
line_1=folium.vector_layers.PolyLine(coords_1,popup='<b>route</b>',tooltip='route',color='blue',weight=5).add_to(f1)
f1.add_to(map5)
folium.LayerControl().add_to(map5)
map5
```

```python
f1=folium.FeatureGroup("route")
#Adding lines to the different feature groups
line_1=folium.vector_layers.PolyLine(coords_1,popup='<b>path of route</b>',tooltip='route',color='blue',weight=9).add_to(f1)
#Adding Markers to the map
folium.Marker(location=[16.780001,80.849998],tooltip='Nuzvid',icon=folium.Icon(color='green',prefix='glyphicon',icon='off')).add_to(map5)
folium.Marker(location=[16.7430141,80.8668691],tooltip='Hanuman Junction',icon=folium.features.CustomIcon('https://www.iconpacks.net/icons/2/free-location-map-icon-2956-thumb.png')).add_to(map5)
folium.Marker(location=[16.703285, 81.100388],tooltip='Eluru',icon=folium.features.CustomIcon('https://www.seekpng.com/png/detail/20-203333_map-png-google-maps-icon-free-icons-red.png')).add_to(map5)
folium.Marker(location=[17.004393,81.783325],tooltip='Rajahmundry',icon=folium.Icon(color='red',prefix='glyphicon',icon='off')).add_to(map5)
folium.Marker(location=[17.0504374,82.1659243],tooltip='Samarlakota',icon=folium.Icon(color='red',prefix='glyphicon',icon='off')).add_to(map5)
folium.Marker(location=[16.989065,82.247467],tooltip='Kakinada',icon=folium.Icon(color='blue',prefix='glyphicon',icon='off')).add_to(map5)
f1.add_to(map5)
folium.LayerControl().add_to(map5)
map5
```
